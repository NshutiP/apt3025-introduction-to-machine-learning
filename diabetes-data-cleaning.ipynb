```python
#Load the diabetes dataset
import pandas as pd
data = pd.read_csv('../data/diabetes.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
# Convert to nulls all the zeros in columns that should not have zeros 
import numpy as np
cols = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
data[cols] = data[cols].replace({0:np.nan})
```

```python
data.head()
```

```python
X = data.drop('Outcome', axis=1)
y = data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
# Calculate medians
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
imputer.fit(X_train)
X_train.median().values
```

```python
# Transform the training data by replacing nulls with medians
X_train_tr = imputer.transform(X_train)	# This returns a numpy array
X_train_tr_df = pd.DataFrame(X_train_tr, columns=X_train.columns, index=X_train.index) # convert np array to pd df
```

```python
X_train_tr_df.info()
```

```python
# Standadize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_tr2 = scaler.fit_transform(X_train_tr_df) # This returns a numpy array
X_train_tr2_df = pd.DataFrame(X_train_tr2, columns=X_train.columns, index=X_train.index)
X_train_tr2_df.head()
```

```python
# Replace nulls with medians in the test set
X_test_tr = imputer.transform(X_test)   # This returns a numpy array
X_test_tr_df = pd.DataFrame(X_test_tr, columns=X_test.columns, index=X_test.index) # convert np array to pd df
X_test_tr_df.info()
```

```python
# Scale the test set
X_test_tr2 = scaler.fit_transform(X_test_tr_df) # This returns a numpy array
X_test_tr2_df = pd.DataFrame(X_test_tr2, columns=X_test.columns, index=X_test.index)
X_test_tr2_df.head()
```

```python
# Train a model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train_tr2_df.values, y_train.values)
```

```python
# Evaluate the model
lr.score(X_test_tr2_df.values, y_test.values)
```
